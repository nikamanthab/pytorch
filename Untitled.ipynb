```python
class DarkData(Dataset):
    def __init__(self,in_dir,out_dir):
        self.in_dir = in_dir
        self.out_dir = out_dir
        os.chdir(in_dir)
        self.in_list = os.listdir()
        os.chdir(out_dir)
        self.out_list = os.listdir()
        print(self.in_list,self.out_list)
        
    def __len__(self):
        return len(self.in_list)

    def __getitem__(self,idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()
        os.chdir(self.in_dir)
        in_name = self.in_list[idx]
        in_image = io.imread(in_name)
        in_image = np.array(in_image)
        in_image = cv2.resize(in_image,(512,512))
        in_image = np.moveaxis(in_image, -1, 0)
        os.chdir(self.out_dir)
        out_name = self.out_list[idx]
        out_image = io.imread(out_name)
        out_image = np.array(out_image)
        out_image = cv2.resize(out_image,(512,512))
        out_image = np.moveaxis(out_image, -1, 0)
        sample = {'in_image': torch.as_tensor(in_image,dtype=torch.float64), 
                  'out_image': torch.as_tensor(out_image,dtype=torch.float64)}
        return sample
```

```python
# Network Architecture
class Network(nn.Module):
    def __init__(self):
        super(Network,self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3,out_channels=4,kernel_size=6,padding=3)
        self.conv2 = nn.Conv2d(in_channels=4,out_channels=8,kernel_size=6,padding=3)
        self.conv3 = nn.Conv2d(in_channels=8,out_channels=16,kernel_size=6,padding=3)
        self.conv4 = nn.Conv2d(in_channels=16,out_channels=8,kernel_size=6,padding=3)
        self.conv5 = nn.Conv2d(in_channels=8,out_channels=4,kernel_size=6,padding=3)
        self.conv6 = nn.Conv2d(in_channels=4,out_channels=3,kernel_size=6,padding=3)
        
#         self.fc1 = nn.Linear(in_features=12*4*4,out_features=120)
#         self.fc2 = nn.Linear(in_features=120,out_features=60)
#         self.out = nn.Linear(in_features=60,out_features=10)
    
    def forward(self,t):
        #forward pass
        print("dim:",t.shape)
        t = self.conv1(t)
        print("dim:",t.shape)
        t = F.relu(t)
        print("dim:",t.shape)
        t = F.max_pool2d(t,kernel_size=2, stride = 1)
        print("dim:",t.shape)
        t = self.conv2(t)
        t = F.relu(t)
        t = F.max_pool2d(t,kernel_size=2, stride = 1)
        
        t = self.conv3(t)
        t = F.relu(t)
        t = F.max_pool2d(t,kernel_size=2, stride = 1)
        
        t = self.conv4(t)
        t = F.relu(t)
        t = F.max_pool2d(t,kernel_size=2, stride = 1)
        
        t = self.conv5(t)
        t = F.relu(t)
        t = F.max_pool2d(t,kernel_size=2, stride = 1)
        
        t = self.conv6(t)
        t = F.relu(t)
        t = F.max_pool2d(t,kernel_size=2, stride = 1)
        
        print(t.shape)
#         -----------------------------------------------------
#         t = t.reshape(-1,12*4*4)
#         t = self.fc1(t)
#         t = F.relu(t)
        
#         t = self.fc2(t)
#         t = F.relu(t)
        
#         t = self.out(t)
#         t = F.softmax(t, dim=1)
        
        return t
```

```python
#training summarized
fcn = Network()
train_loader = torch.utils.data.DataLoader(darkdataset,batch_size=100)
optimizer = optim.Adam(fcn.parameters(),lr=0.001)

epoch = 0
no_of_epoch = 1000
#batch = next(iter(train_loader))
for epoch in range(no_of_epoch):
    total_loss = 0
    total_correct = 0
    for batch in train_loader:
        dark_imgs,light_imgs = batch['in_image'],batch['out_image']
#         print(dark_imgs)
        preds = fcn(dark_imgs.float())
        loss = loss = F.mse_loss(preds,light_imgs.float())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss+=loss.item()
#         total_correct+=get_num_correct(preds,labels)

    print("epoch:",epoch,"total_loss:",total_loss)
```
